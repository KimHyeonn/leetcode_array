```python
from bs4 import BeautifulSoup
```

```python
html_doc = """

<html><head><title>The Dormouse's story</title></head>

<body>

<p class="title"><b>The Dormouse's story</b></p>



<p class="story">Once upon a time there were three little sisters; and their names were

<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,

<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and

<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;

and they lived at the bottom of a well.</p>



<p class="story">...</p>

"""
```

```python
soup = BeautifulSoup(html_doc, 'html.parser') # html 형식으로 파싱하겠다.
```

```python
# BeautifulSoup을 이용해 파싱한 후, prettify() 매서드를 이용해 구조화하여 살펴본다.

print(soup.prettify())
```

```python
# BeautifulSoup 인스턴스 매서드들을 통해 필요 정보 추출

## 1. find() & find_all()
## - 기본적으로 tag 이름을 전달하여 매칭되는 태그의 첫번째 정보를 추출할 수 있다.
## <a..> tag에 자매들의 이름이 있으므로, a 태그에 접근해보자.

soup.find('a')
```

```python
soup.find_all('a')
```

```python
type(soup.find('a'))
```

```python
# find()로 반환된 결과가 bs4 Tag 객체임을 알 수 있다. (문자열 x)
# 그러므로 해당 객체의 특정 매서드나 속성이 있을 것으로 예상할 수 있다.
# 태그 내 정보, 즉 자매 이름에 관심이 있으므로 이를 가져오는 방법은

soup.find('a').string
# sting 속성에 해당 태그의 텍스트 정보가 담겨있다.
```

```python
type(soup.find_all('a'))
```

```python
# find_all()로 반환된 결과는 Result set
# find()와 달리 bs4 객체들을 리스트로 주고있다는 점. -> list comprehension 적용

[x.string for x in soup.find_all('a')]
```

```python
## 2. select()
## - select()는 css selector 를 이용하거나, 태그 내의 특정 클래스명을 이용하여 데이터를 가져올 때 사용한다.
## - 특정 태그로 접근도 가능하며, 해당 태그 내의 하위 태그로도 확대 적용 가능하다.

## html_doc 에서 p태그의 'story' 클래스의 텍스트 데이터를 추출해보자.

soup.select('.story')[0].text
```

```python
## find() 이용시

soup.find('p', class_ = 'story')
```

```python
## find() 이용시, error 케이스

soup.find('p', class_= 'story')[0]
```

```python
[i.string for i in soup.find('p', class_='story')][0]
```

```python
## select() 이용, 하위 태그 접근

soup.select('p > a')
```

```python
## quiz. 세 자매 이름을 select()로 추출

[x.string for x in soup.select('p > a')]
```

```python
charta = []
for i in range(1,101):
    b = [x.string for x in song_list1[i].select('span > a')]
    if len(b) == 2:
        charta.append([x.string for x in song_list1[i].select('span > a')])
    elif len(b) == 3:
        a = [x.string for x in song_list1[i].select('span > a')][1]+\
        ' '+[x.string for x in song_list1[i].select('span > a')][2]
        charta.append([[x.string for x in song_list1[i].select('span > a')][0], a])
    elif len(b) == 4:
        a = [x.string for x in song_list1[i].select('span > a')][1]+\
        ' '+[x.string for x in song_list1[i].select('span > a')][2]+\
        ' '+[x.string for x in song_list1[i].select('span > a')][3]
        charta.append([[x.string for x in song_list1[i].select('span > a')][0], a])
    elif len(b) == 5:
        a = [x.string for x in song_list1[i].select('span > a')][1]+\
        ' '+[x.string for x in song_list1[i].select('span > a')][2]+\
        ' '+[x.string for x in song_list1[i].select('span > a')][3]+\
        ' '+[x.string for x in song_list1[i].select('span > a')][4]
        charta.append([[x.string for x in song_list1[i].select('span > a')][0], a])
charta
```

```python
pd.DataFrame(charta, columns = ['song','artist'])
```
